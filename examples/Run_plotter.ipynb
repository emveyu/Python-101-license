```python
# UPLC plot
import plotter as pl
import numpy as np

x,y,filenm = pl.load_tool.txt_load(start_text='Wavelength(nm)	280', spacer='	', wait=3)
ny = np.array(y)/1000
pl.plot_tool.curve(x,ny,
                  title='Wavelength 280 (nm)', tfontsize=10,
                  xlabel='time (min)', xfontsize=10,
                  ylable='intensity', yfontsize=10,
                  linewidth=1,
                  linecolor='blue', #'red', 'blue' or more in https://www.w3schools.com/colors/colors_names.asp
                  linestyle='-', #':' dotted line; '--' dash line
                  width=5,
                  height=3,
                  filenm=filenm[:-4]
                 )
```

```python
# clMS chromatography plot
import plotter as pl
import numpy as np
x,y,filenm = pl.load_tool.txt_load(spacer='	')
pl.plot_tool.curve(x,y,
                  title='chromatography', tfontsize=10,
                  xlabel='time (min)', xfontsize=10,
                  ylable='intensity', yfontsize=10,
                  linewidth=1,
                  linecolor='blue', #'red', 'blue' or more in https://www.w3schools.com/colors/colors_names.asp
                  linestyle='-', #':' dotted line; '--' dash line
                  width=5,
                  height=3,
                  filenm=filenm[:-4]+'-chrom'
                 )
```
